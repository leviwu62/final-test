```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title("Sin Wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.plot(x,y)
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.cos(x*np.pi/180)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.title("Sin Wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.plot(x,y)
pt.show()
```
